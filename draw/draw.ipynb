```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import savgol_filter

# 读取csv文件
df = pd.read_csv('mpe_mpe-large-spread-v1_ac_att.csv')

# 设置图的大小
plt.figure(figsize=(10,6))

# 平滑处理的窗口大小和多项式次数
window_size = 21
poly_order = 3

# 平滑处理数据
smoothed_data = {
    'JAMMAS_4': savgol_filter(df['mpe_mpe-large-spread-v1_ac_att_JAMMAS_4 - episode_return'], window_size, poly_order),
    'JAMMAS_3': savgol_filter(df['mpe_mpe-large-spread-v1_ac_att_JAMMAS_3 - episode_return'], window_size, poly_order),
    'JAMMAS_1': savgol_filter(df['mpe_mpe-large-spread-v1_ac_att_JAMMAS_1 - episode_return'], window_size, poly_order),
    'JAMMAS_2': savgol_filter(df['mpe_mpe-large-spread-v1_ac_att_JAMMAS_2 - episode_return'], window_size, poly_order)
}

# 绘制平滑处理后的四条折线（0-1000范围）
plt.plot(df['epoch'][:1000], smoothed_data['JAMMAS_4'][:1000], label='Smoothed JAMMAS_4', color="red")
plt.plot(df['epoch'][:1000], smoothed_data['JAMMAS_3'][:1000], label='Smoothed JAMMAS_3', color="green")
plt.plot(df['epoch'][:1000], smoothed_data['JAMMAS_1'][:1000], label='Smoothed JAMMAS_1', color="blue")
plt.plot(df['epoch'][:1000], smoothed_data['JAMMAS_2'][:1000], label='Smoothed JAMMAS_2', color="yellow")

# 绘制之前的线，并降低透明度（0-1000范围）
plt.plot(df['epoch'][:1000], df['mpe_mpe-large-spread-v1_ac_att_JAMMAS_4 - episode_return'][:1000], alpha=0.3, label='JAMMAS_4 (Original)', color="red")
plt.plot(df['epoch'][:1000], df['mpe_mpe-large-spread-v1_ac_att_JAMMAS_3 - episode_return'][:1000], alpha=0.3, label='JAMMAS_3 (Original)', color="green")
plt.plot(df['epoch'][:1000], df['mpe_mpe-large-spread-v1_ac_att_JAMMAS_1 - episode_return'][:1000], alpha=0.3, label='JAMMAS_1 (Original)', color="blue")
plt.plot(df['epoch'][:1000], df['mpe_mpe-large-spread-v1_ac_att_JAMMAS_2 - episode_return'][:1000], alpha=0.3, label='JAMMAS_2 (Original)', color="yellow")

# 设置图例
plt.legend()

# 设置标题和坐标轴标签
plt.title('Smoothed Return Values over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Smoothed Return Value')

# 显示图形
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import savgol_filter

# 读取csv文件
df = pd.read_csv('mpe_mpe-large-spread-v1_ac_att.csv')

# 设置图的大小
plt.figure(figsize=(10, 6))

# 平滑处理的窗口大小和多项式次数
window_size = 16
poly_order = 3

# 平滑处理数据
smoothed_data = {
    '_4': savgol_filter(df['mpe_mpe-large-spread-v1_ac_att_JAMMAS_4 - episode_return'], window_size, poly_order),
    '_3': savgol_filter(df['mpe_mpe-large-spread-v1_ac_att_JAMMAS_3 - episode_return'], window_size, poly_order),
    '_1': savgol_filter(df['mpe_mpe-large-spread-v1_ac_att_JAMMAS_1 - episode_return'], window_size, poly_order),
    '_2': savgol_filter(df['mpe_mpe-large-spread-v1_ac_att_JAMMAS_2 - episode_return'], window_size, poly_order)
}

# 定义一组更好看的颜色
colors = ['dodgerblue', 'orange', 'green', 'red']


for idx, key in enumerate(smoothed_data.keys()):
    plt.plot(df['epoch'][:1000], smoothed_data[key][:1000],label=f'Mpe_att_head_{key}(smoothed)', color=colors[idx])

# 绘制之前的线，并降低透明度（0-1000范围）
for idx, key in enumerate(smoothed_data.keys()):
    plt.plot(df['epoch'][:1000], df[f'mpe_mpe-large-spread-v1_ac_att_JAMMAS{key} - episode_return'][:1000], alpha=0.3, label=f'Mpe_att_head_{key} (Original)', color=colors[idx])

# 设置图例
plt.legend()

# 设置标题和坐标轴标签
plt.title('Mpe-large-spread-v1 Smoothed Return Values over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Smoothed Return Value')

# 显示图形
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import savgol_filter

# 读取csv文件
df = pd.read_csv('ac_noise.csv')

# 设置图的大小
plt.figure(figsize=(10, 6))

# 平滑处理的窗口大小和多项式次数
window_size = 16
poly_order = 3

# 平滑处理数据
smoothed_data = {
    'Att_Noise': savgol_filter(df['mpe_mpe-large-spread-v1_ac_att_noise_JAMMAS_1 - episode_return'], window_size, poly_order),
    'Att_Normal': savgol_filter(df['mpe_mpe-large-spread-v1_ac_att_JAMMAS_1 - episode_return'], window_size, poly_order),
}


# 绘制平滑处理后的四条折线
plt.plot(df['epoch'][:1000], smoothed_data['Att_Noise'][:1000], label='Smoothed Att_Noise', color="dodgerblue")
plt.plot(df['epoch'][:1000], smoothed_data['Att_Normal'][:1000], label='Smoothed Att_Normal', color="orange")

# 绘制之前的线，并降低透明度
plt.plot(df['epoch'][:1000], df['mpe_mpe-large-spread-v1_ac_att_noise_JAMMAS_1 - episode_return'][:1000], alpha=0.3, label='Att_Noise (Original)', color="dodgerblue")
plt.plot(df['epoch'][:1000], df['mpe_mpe-large-spread-v1_ac_att_JAMMAS_1 - episode_return'][:1000], alpha=0.3, label='Att_Normal (Original)', color="orange")

# 设置图例
plt.legend()

# 设置标题和坐标轴标签
plt.title('Smoothed Return Values over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Smoothed Return Value')

# 显示图形
plt.show()
```
